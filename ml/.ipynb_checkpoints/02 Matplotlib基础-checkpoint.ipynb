# Matplotlib 基础

```python
import matplotlib.pyplot as plt
```

```python
import numpy as np
x = np.linspace(0, 10, 100) #在[0,10]区间内生成等距离100个数
x
```

```python
y = np.sin(x)
```

```python
plt.plot(x, y)
plt.plot(x, np.cos(x), color='red', linestyle='--')
plt.show()
```

```python
plt.plot(x, y)
plt.plot(x, np.cos(x), color='red', linestyle='--')
plt.axis([-1, 11, -2, 2]) #设置数轴范围
plt.xlabel('x')
plt.ylabel('y')
plt.show()
```

```python
plt.plot(x, y, label='sin(x)')
plt.plot(x, np.cos(x), color='red', linestyle='--', label='cos(x)')
plt.axis([-1, 11, -2, 2])
plt.xlabel('x')
plt.ylabel('y')
plt.legend() #加图示
plt.title('sin & cos') #加标题
plt.show()
```

```python
x = np.random.normal(0,1,10000)
y = np.random.normal(0,1,10000)
plt.scatter(x, y, alpha=0.1) #散点图, alpha为透明度
plt.show()
```
