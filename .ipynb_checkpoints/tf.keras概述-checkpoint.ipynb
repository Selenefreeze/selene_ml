```python
import pandas as pd
```

```python
data = pd.read_csv('./dataset/Salary_Data.csv')
```

```python
data
```

```python
import matplotlib.pyplot as plt
```

```python
#散点图-scatter
plt.scatter(data.YearsExperience, data.Salary)
```

```python
x = data.YearsExperience
y = data.Salary
```

```python
import tensorflow as tf
```

```python
print('Tensorflow Version:{}'.format(tf.__version__) )
```

```python
#序贯模型（Sequential):单输入单输出，一条路通到底，层与层之间只有相邻关系，没有跨层连接。
#编译速度快，操作比较简单。
modle = tf.keras.Sequential()
```

```python
#dense:全连接层  相当于添加一个层(输出维度，输入的形式=元组)
modle.add(tf.keras.layers.Dense(1, input_shape=(1,)))
```

```python
modle.summary() # ax + b
```

```python
modle.compile(optimizer='adam',
             loss='mse'
             )
#adam = 随机梯度下降
#mse = 均方误差
```

```python
#训练过程
history = modle.fit(x, y, epochs=5000)
```

```python
modle.predict(x)
```

```python
#x的格式为pd.Series
modle.predict(pd.Series([12]))
```
